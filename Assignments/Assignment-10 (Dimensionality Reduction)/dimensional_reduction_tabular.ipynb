```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA, IncrementalPCA, FactorAnalysis
from sklearn.manifold import TSNE, MDS, Isomap
from sklearn.preprocessing import StandardScaler
import umap
from sklearn.datasets import load_wine
from sklearn.manifold import LocallyLinearEmbedding

# 1. **Load Wine Dataset**
# The Wine dataset contains 178 samples, each with 13 features.
wine = load_wine()
X = wine.data  # Features
y = wine.target  # Labels
target_names = wine.target_names  # Target class names

# 2. **Preprocess the Data**
# Standardizing the data (scaling features to have zero mean and unit variance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. **PCA Preprocessing for t-SNE and UMAP**
# Reduce dimensions using PCA before t-SNE or UMAP to speed up the process
pca = PCA(n_components=2)  # Reduce to 2 dimensions before t-SNE and UMAP
X_pca = pca.fit_transform(X_scaled)
```

```python
## LLE (Locally Linear Embedding)
lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2)
X_lle = lle.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('LLE - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## t-SNE (t-Distributed Stochastic Neighbor Embedding)
tsne = TSNE(n_components=2, random_state=42, verbose=1)
X_tsne = tsne.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('t-SNE - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## ISOMAP
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('Isomap - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## UMAP (Uniform Manifold Approximation and Projection)
umap_model = umap.UMAP(n_components=2)
X_umap = umap_model.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('UMAP - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## MDS (Multidimensional Scaling)
mds = MDS(n_components=2, random_state=42)
X_mds = mds.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('MDS - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## PCA (Principal Component Analysis)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('PCA - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## Kernel PCA
kpca = KernelPCA(n_components=2, kernel='rbf')
X_kpca = kpca.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('Kernel PCA - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## Incremental PCA
ipca = IncrementalPCA(n_components=2)
X_ipca = ipca.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_ipca[:, 0], X_ipca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('Incremental PCA - Wine Dataset')
plt.colorbar()
plt.show()
```

```python
## Factor Analysis
factor_analysis = FactorAnalysis(n_components=2)
X_factor = factor_analysis.fit_transform(X_scaled)

# Plot the result
plt.figure(figsize=(8, 6))
plt.scatter(X_factor[:, 0], X_factor[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('Factor Analysis - Wine Dataset')
plt.colorbar()
plt.show()
```
